### `.pivot_table()`
---
A função `.pivot_table()` funciona como uma tabela dinâmica do excel. Por meio dela, podemos resumir os dados de uma maneira que facilite a compreensão.

Ela possui 3 argumentos principais: `values`, `index`, `columns`

O primeiro recebe a coluna numérica do dataframe original que queremos resumir, no nosso caso, queremos analisar a *quantidade produzida* .

O segundo argumento controla qual coluna será utilizada como índice da tabela. Como nós queremos saber, para cada ano, a

```python
df_pivot = df.pivot_table(values = 'quantidade_produzida', index = 'id_municipio', columns = 'ano')
```
