```python
import pandas as pd 
from pandas_datareader import data as web 
import plotly.graph_objs as go
```

<br>

<br>

Com o pandas_datareader, n√≥s conseguimos importar dados de fontes como a **<font color = #055e9e > OECD, World Bank, Nasdaq** <font color = black> entre outros. 
<br>

No, nosso caso, selecionamos os dados do Yahoo. Ao selecionar os dados do Yahoo, conseguimos selecionar o ativo que desejamos analisar bem como o per√≠odo da an√°lise. Ap√≥s isso ele nos retorna as informa√ß√µes di√°rias do ativo como, por exemplo, pre√ßo de abertura e fechamento. 

<br>

<br>

```python
df = web.DataReader(f'PETR4.SA', data_source = 'yahoo', start = f'03-10-2021', end = '02-20-2022')
display(df)
```

<br>

<br>

**<font color = #055e9e >Com a boblioteca do Plotly n√≥s conseguimos visualizar esses dados de v√°rias maneiras, uma delas √© no formato de Candlesticks** <font color = black >. Estes estilos de gr√°ficos s√£o comumente utilizados para representar o desempenho de a√ß√µes no mercado financeiro . 
<br>

Voc√™ somente precisa passar algumas informa√ß√µes para a realiza√ß√£o dos gr√°ficos de Candlestick. Elas s√£o o pre√ßo de abertura, fechamento e valor m√°ximo e m√≠nimo do ativo do dia. Depois √© s√≥ customizar o seu gr√°fico da maneira como desejar ! üòÅ

<br>

<br>

```python
df['Date'] = pd.to_datetime(df.index)
df = df.set_index('Date')
```

```python
fig = go.Figure(data=[go.Candlestick(x = df.index,
                                     open = df['Open'],
                                     high = df['High'],
                                     low = df['Low'],
                                     close = df['Close'])])
fig.update_layout(
    title = 'PETR4, 03/2021 - 02/2022',
    yaxis_title = 'Valor do Ativo',
    xaxis_title = 'Data',
    template = "plotly_white",
    xaxis_rangeslider_visible=False)
```
